```python
import re

def clean_data(X : str) -> str:
    X = X.lower().strip()
    X = re.sub(r'[^a-zA-Z0-9]', ' ', X)
    
    return X
```

```python
for text in df_test['text']:
    df_test['text'] = clean_data(text)
    
```

```python
print(df_test['text'][0])
```

```python
text_array = np.array(df_test['text'])
target_array = np.array(df_test['target'])
```

```python
text_array.shape 
text_array[0]
```

```python
text_array.shape
```

```python
X_train , X_test , y_train , y_test = text_array[:5000] , text_array[5000:] , target_array[:5000] , target_array[5000:]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()

def processed_data(X : str):
    return vectorizer.fit_transform(X)

```

```python
X_train_matrix = processed_data(X_train)
X_test_matrix = processed_data(X_test)
     
```

```python
from sklearn.linear_model import LogisticRegression

logi_reg = LogisticRegression()
logi_reg.fit(X_train_matrix , y_train)

```

```python
from sklearn.model_selection import cross_val_predict , cross_val_score

score = cross_val_score(logi_reg , X_train_matrix , y_train , cv=3 , scoring="accuracy")
```
